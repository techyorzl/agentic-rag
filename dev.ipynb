```python
import os
import json
import requests
from dotenv import load_dotenv

load_dotenv()

api_key = os.getenv("SERP_API_KEY")
if not api_key:
    raise ValueError("Please set the SERP_API_KEY environment variable.")
```

```python
query = "lithium battery"

url = f"https://serpapi.com/search?engine=google_patents&q={query}&api_key={api_key}"
```

```python
resp = requests.get(url)
if resp.status_code == 200:
    data = resp.json()
    
    patents = data.get("organic_results", [])

    with open("files/patents.json", "w") as f:
        json.dump(patents, f, indent = 4)
```

```python
url = "https://serpapi.com/search.json?engine=google_patents_details&patent_id=patent%2FCN109659487B%2Fen&api_key=" + api_key
resp = requests.get(url)
if resp.status_code == 200:
    patent_details = resp.json()
    with open("files/patent_det.json", "w") as f:
        json.dump(patent_details, f, indent=4)
```

```python
dir_files = os.listdir("results")
for file in dir_files:
    if file.endswith(".json"):
        with open(os.path.join("results", file), "r") as f:
            content = json.load(f)
            print(content.get("abstract", ""))
            print("\n" + "="*40 + "\n")
```

```python
emb_url = "http://localhost:11434/api/embeddings"
headers = {
    "Content-Type": "application/json"
}
data = {
    "model": "nomic-embed-text",
    "prompt": "the sky is blu because of rayleigh scattering"
}

response = requests.post(emb_url, headers=headers, json=data)
if response.status_code == 200:
    embedding = response.json()
    embed = embedding.get("embedding", [])
    print(len(embed))
```
